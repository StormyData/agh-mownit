##### 1. Przygotowanie wzorców

Proces rozpoznawania znaków zacząłem od wygenerowania wzorców, poprzez narysowanie po koleji każdego rozpoznawanego znaku, odszumienie go jak w kroku 4. i wykonanie konwolucji z wyrenderowanym, odszumionym tekstem zawierającym wszystkie rozpoznawane znaki, następnie maksymalna wartość w konwolucji jest zapamiętywana jako granica detekcji do wykorzystania później, wartości tego kroku są cache-owane przy pierwszym uruchomieniu

##### 2. Generowanie szerokości odstępów potrzebnych do detekcji spacjii

Wyznaczana jest najpierw długość spacji, poprzez odjęcie odległości między znakami oddzielonymi pojedyńczą spacją od odległości między znakami oddzielonymi podwójną spacją, wartości tego kroku są cache-owane przy pierwszym uruchomieniu

Na koniec wyznaczane są odległości między każdą parą rozpoznawanych znaków rozdzielonych spacją

##### 3. Wstępne przetwarzanie: obrót

Na obrazie wejściowym uruchamiamy algorytm hough-a do rozpoznawania lini w celu wyznaczenia kąta, o który należy obrócić obraz

następnie obraz jest obracany o wyznaczony wcześniej kąt

##### 4. Wstępne przetwarzanie: filtrowanie

Na obróconym w poprzednim kroku obrazie wykonywane jest filtrowanie z pomocą dekompozycji SVD odrzucające 87.5% najmniej znaczących wektorów własnych


##### 5. Rozpoznawanie znaków: detekcja znaków

Obraz z kroku 4. konwolujemy po koleji z każdym wzorcem wygenerowanym w kroku 1. następnie zapisujemy współrzędne punktów, dla których konwolucja przekroczyła wartość graniczną (również wyznaczoną w kroku 1.)

współrzędne te w połączeniu z wielkością wzorca tworzą prostokąty, określające miejsca, gdzie znaleziono dany znak

następnie prostokąty nachodzące na siebie są łączone w jeden

i dla każdego znaku powstaje lista prostokątów w którch znaleziono dany znak

##### 6. Rozpoznawanie znaków: wykluczanie znaków znalezionych w innych

Dla każdego prostokąta, dla każdego znaku sprawdzane jest jak bardzo zachodzi on na prostokąt innego znaku, i jeżeli znak ma wyższy priorytet i stopień zachodzenia przekracza wcześniej ustalony próg dany prostokąt zostaje odrzucony

##### 7. Rozpoznawanie wierszy

z połorzeń i wysokości każdego znaku generowane są przedziały pionowe, których rozłączne części będą stanowiły kolejne linie wykrywanego tekstu, trzeba dodać, że ten sposób wykrywania lini nie zadziała, jeżeli rzut na oś y znaków z lini niżej będzie się przecinał z rzutem znaków z lini wyżej
